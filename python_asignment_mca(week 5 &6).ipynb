Single File Programming Question
Problem Statement
Implement a program that demonstrates the concept of aggregation using classes. Define two classes: Salary and EmployeeOne.

Salary Class contains attributes pay (monthly salary) and bonus, and a method annual_salary() that calculates and returns the annual salary by multiplying the monthly pay by 12 and adding the bonus.



EmployeeOne Class represents an employee with attributes name, age, and agg_salary (an instance of Salary class). It has a method total_sal() that returns the total annual salary of the employee by calling the annual_salary() method of the Salary object.

Input format :
The first line of input consists of an integer S, representing the monthly salary of the employee.

The second line consists of an integer B, representing the annual bonus received by the employee.

The third line consists of a string N, representing the name of the employee.

The fourth line consists of an integer A, representing age of the employee.

Output format :
The output prints the total annual salary of the employee calculated as (pay * 12) + bonus.

Refer to the sample output for formatting specifications.

Code constraints :
In this scenario, the given test cases will fall under the following constraints:

5000 ≤ S ≤ 100000

2000 ≤ B ≤ 25000

Sample test cases :
Input 1 :
5000
2000
John Doe
30
Output 1 :
62000
Input 2 :
4000
1500
Jane Smith
28
Output 2 :
49500
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
# You are using Python
class Salary:
    def __init__(self,pay,bonus):
        self.pay=pay
        self.bonus=bonus
    def annual_salary(self):
        return (self.pay*12)+self.bonus
class EmployeeOne:
    def __init__(self,name,age,agg_salary):
        self.name=name
        self.age=age
        self.agg_salary=agg_salary
    def total_sal(self):
        return self.agg_salary.annual_salary()
S=int(input().strip())
B=int(input().strip())
N=input().strip()
A=int(input().strip())
salary_instance=Salary(S,B)
employee=EmployeeOne(N,A,salary_instance)
print(employee.total_sal())
```

Implement a program that demonstrates aggregation and association between a Student and Department class. The Student class generates a registration number based on its ID and a department ID, while the Department class associates with a Student object to retrieve and print the registration number.



Attributes: 

Student Class:

id: Represents the student ID.
Department Class:

id: Represents the department ID.
student: Represents an instance of the Student class.
Input format :
The first line of input consists of a string I, representing the student ID.

The second line consists of a string D, representing the department ID.

Output format :
The output prints a single string which is the student registration number in the format: "studentID-departmentID".

The program will be evaluated only after the “Submit Code” is clicked.

```python
# You are using Python
class Student:
    def __init__(self,stud_id):
        self.id=stud_id
    def get_registration_no(self,dept_id):
        return f"{self.id}-{dept_id}"
class Department:
    def __init__(self,dept_id):
        self.id=dept_id
        self.student=None
    def associate_stud(self,student):
        self.student=student
    def print_reg_no(self):
        if self.student:
            reg_no=self.student.get_registration_no(self.id)
            print(reg_no)
stud_id=input().strip()
dept_id=input().strip()
student=Student(stud_id)
department=Department(dept_id)
department.associate_stud(student)
department.print_reg_no()
            
```

Problem Statement



Lisa loves building toy houses using blocks. She has a special set of blocks that she uses to create houses. Each block has a certain length, width, and height. Lisa is curious about the total surface area of the toy house she's building, excluding the roof. Help her with the code to calculate it.



Implement a program that has a class named House and takes the dimensions of a toy house (length, width, and height) and calculates the total surface area (TSA) of the house without the roof.



Formula: TSA = 2 * [(Width * Height) + (Width * Height) + (Height * Length) + (Height * Length)]

Input format :
The input consists of three float values in each line: length, width, and height of the toy house.

Output format :
The output displays a float value - the total surface area of the toy house (excluding the roof) rounded off to two decimal places followed by " square units.".



Refer to the sample output for the formatting specifications.

Code constraints :
1 ≤ length, width, height ≤ 1000

Sample test cases :
Input 1 :
5.5
4.2
3.8
Output 1 :
147.44 square units.
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
# You are using Python
class House:
    def __init__(self,length,width,height):
        self.length=length
        self.width=width
        self.height=height
    def TSA(self):
        tsa=2*((self.width*self.height)+(self.width*self.height)+(self.height*self.length)+(self.height*self.length))
        return tsa
l=float(input())
w=float(input())
h=float(input())
h=House(l,w,h) 
tsa=h.TSA()
print(f"{tsa:.2f} square units.")
```

Problem Statement



Imagine you are developing a text analysis tool for a cybersecurity company. Your task is to create a function that analyzes input strings to categorize and count the characters into four categories: uppercase letters, lowercase letters, digits, and special characters. The company needs this tool to process log files and identify potential security threats.



Function Signature: analyze_string(input_string)

Input format :
The input consists of a single string (without space), which may include uppercase letters, lowercase letters, digits, and special characters.

Output format :
The first line contains an integer representing the count of uppercase letters in the format "Uppercase letters: [count]".

The second line contains an integer representing the count of lowercase letters in the format "Lowercase letters: [count]".

The third line contains an integer representing the count of digits in the format "Digits: [count]".

The fourth line contains an integer representing the count of special characters in the format "Special characters: [count]".



Refer to the sample output for the formatting specifications.

Code constraints :
In this scenario, the given test cases will fall under the following constraints:

1 ≤ Length of the input string ≤ 100

Sample test cases :
Input 1 :
Hello123
Output 1 :
Uppercase letters: 1
Lowercase letters: 4
Digits: 3
Special characters: 0
Input 2 :
12345
Output 2 :
Uppercase letters: 0
Lowercase letters: 0
Digits: 5
Special characters: 0
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
# You are using Python
def analyze_string(input_string):
    uc=0
    lc=0
    dc=0
    sc=0
    for char in input_string:
        if char.isupper():
            uc+=1
        elif char.islower():
            lc+=1
        elif char.isdigit():
            dc+=1
        else:
            sc+=1
    print(f'Uppercase letters: {uc}')
    print(f'Lowercase letters: {lc}')
    print(f'Digits: {dc}')
    print(f'Special characters: {sc}')
in1=input()
analyze_string(in1)
```

Single File Programming Question
Problem Statement



Sophia manages payroll at a local business where employees are paid based on the number of hours they work each month. She needs to calculate the monthly salary for each employee, considering a fixed hourly rate and standard deductions. The deductions amount to 15% of the gross salary.



Help Sophia write a program that calculates the net monthly salary for each employee. The program should read the total hours worked and the hourly rate from the input, compute the gross salary, apply the deductions, and then output the net salary.



Example

Input

40

12.50



Output 

Hours Worked: 40.0 hours

Hourly Rate: 12.50 per hour

Net Salary: 425.00



Explanation

The gross salary is calculated as 40.0 hours * 12.50/hour = 500.0

The deductions are 15% of the gross salary, which is 0.15 * 500.00 = 75.00

The net salary is gross - deductions = 500.0 - 75.0 = 425.00.



Function Signature: monthly_salary(hours_worked, hourly_rate)

Input format :
The first line contains a positive floating-point number representing the number of hours worked by the employee.

The second line contains a positive floating-point number representing the hourly rate for the employee.

Output format :
The first line displays the number of hours worked in the format "Hours Worked: {hours} hours," where {hours} is the actual number of hours worked with two decimal places.

The second line displays the hourly rate in the format "Hourly Rate: {rate} per hour," where {rate} is the actual hourly rate rounded to two decimal places.

The third line displays the net salary in the format "Net Salary: {salary}," where {salary} is the actual net salary rounded to two decimal places.



Refer to the sample output for the formatting specifications.

Code constraints :
In this scenario, the given test cases will fall under the following constraints:

1 ≤ Number of hours worked ≤ 200

10.0 ≤ Hourly rate ≤ 50.0

The hourly rate is specified with a precision of up to two decimal places.

Sample test cases :
Input 1 :
40.00
12.50
Output 1 :
Hours Worked: 40.00 hours
Hourly Rate: 12.50 per hour
Net Salary: 425.00
Input 2 :
80.00
30.75
Output 2 :
Hours Worked: 80.00 hours
Hourly Rate: 30.75 per hour
Net Salary: 2091.00
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
# You are using Python
def monthly_salary(hours_worked,hourly_rate):
    gross_salary=hours_worked*hourly_rate
    deductions=0.15*gross_salary
    net_salary=gross_salary-deductions
    print(f"Hours Worked: {hours_worked:.2f} hours")
    print(f"Hourly Rate:{hourly_rate:.2f} per hour")
    print(f"Net Salary: {net_salary:.2f}")
hours_worked=float(input())
hourly_rate=float(input())
monthly_salary(hours_worked,hourly_rate)
```
