# Single File Programming Question
**Problem Statement**



Jake is working on a data analysis project where he needs to compute various statistical measures for a given dataset of integer values. The dataset consists of several numbers representing different data points. 



Write a program to calculate and display the mean, median, mode, variance, and standard deviation of the dataset. The program should take a comma-separated list of integers as input and output the calculated statistics.

Input format :
The first line of input consists of a single line containing a comma-separated list of integers representing the dataset.

Output format :
The first line of output prints "Dataset: "[elements].

The second line of output prints "Mean: " followed by the mean of the dataset, formatted to two decimal places.

The third line of output prints "Median: " followed by the median of the dataset, formatted to two decimal places.

The fourth line of output prints "Mode: " followed by the mode of the dataset.

The fifth line of output prints "Variance: " followed by the variance of the dataset, formatted to two decimal places.

The sixth line of output prints "Standard Deviation: " followed by the standard deviation of the dataset, formatted to two decimal places.



Refer to the sample output for the formatting specifications.

Code constraints :
The given test cases will fall under the following constraints:

1 ≤ dataset ≤ 100

Sample test cases :
Input 1 :
10,20,30,40,50
Output 1 :
Dataset: [10, 20, 30, 40, 50]
Mean: 30.00
Median: 30.00
Mode: 10
Variance: 250.00
Standard Deviation: 15.81
Input 2 :
3,5,7,11,15
Output 2 :
Dataset: [3, 5, 7, 11, 15]
Mean: 8.20
Median: 7.00
Mode: 3
Variance: 23.20
Standard Deviation: 4.82
Note :

```python
# You are using Python
import statistics as s
data_in=input().strip()
dataset=list(map(int,data_in.split(',')))
mean=s.mean(dataset)
median=s.median(dataset)
try:
    mode=s.mode(dataset)
except s.StatisticsError:
    mode='NO MODE'
variance=s.variance(dataset)
std_dev=s.stdev(dataset)
print(f'Dataset: {dataset}')
print(f'Mean: {mean:.2f}')
print(f'Median: {median:.2f}')
print(f'Mode: {mode}')
print(f'Variance: {variance:.2f}')
print(f'Standard Deviation: {std_dev:.2f}')
```

# Single File Programming Question
**Problem Statement**



Emma, a software tester, is validating server availability on multiple ports. She needs to write a Python script where a client attempts to connect to a server on different ports and prints whether the connection was successful or not. If a connection fails, the program should display an appropriate error message. 



Help her to solve this problem by writing an efficient Python script that handles multiple connection attempts gracefully.



Note: In this compiler, port 8080 is connected by default, which is why the connection to port 8080 succeeds.

Input format :
The first line of input consists of a string, representing the IP address the client tries to connect to.

The second line consists of a list of integers, representing the ports the client attempts to connect on.

Output format :
For each port, print "Connected to port X" if the connection is successful.

If the connection fails, print "Connection failed on port X: timed out" for time out.

print "Connection failed on port X: Connection refused" if the connection is refused

print "Connection failed on port X: [Errno -2] Name does not resolve" if socket error occurs.



Refer to the sample output for formatting specifications.

Code constraints :
The given test cases fall under the following specifications:

1 ≤ Port Number ≤ 65535

Sample test cases :
Input 1 :
127.0.0.1
8080 9090 10000
Output 1 :
Connected to port 8080
Connection failed on port 9090: Connection refused
Connection failed on port 10000: Connection refused
Input 2 :
256.256.256.256
8080 9090
Output 2 :
Connection failed on port 8080: [Errno -2] Name does not resolve
Connection failed on port 9090: [Errno -2] Name does not resolve
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
import socket
import sys

def check_connection(ip,ports):
    for port in ports:
        try:
            s=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
            s.settimeout(5)
            s.connect((ip,port))
            print(f"Connected to port {port}")
            s.close()
        except socket.timeout:
            print(f"connection failed on port{port}:timed out")
        except socket.error as e:
            if e.errno==-2:
                print(f"connection failed on port {port}: [Errno 2] Name does not resolve")
            else:
                print(f"connection failed on port {port}: Connection refused")
                        
if __name__=="__main__":
    ip_address=input().strip()
    ports=list(map(int,input().split()))
    
    check_connection(ip_address,ports)
```

#Single File Programming Question
**Problem Statement**



Bob, a cybersecurity analyst, needs to check which ports on a given IP address are open, closed, or timing out. This helps him assess security vulnerabilities and ensure only necessary services are running. 



Your task is to implement a port scanner that connects to multiple ports on a target IP address and reports their status. 



Help Bob secure his network!



Note: In this compiler, only port 8080 will be OPEN by default. 

Input format :
The first line of input consists of a string (ip_address), representing the target IP address to be scanned.

The second line consists of space-separated integers (ports[]), representing the port numbers to be scanned.

Output format :
For each port, print "Port X: OPEN" if the connection is successful.

Print "Port X: CLOSED" if the connection is refused.

Print "Port X: TIMEOUT (Possibly closed)" if the connection times out.

If the IP address is invalid, print "Invalid IP address." and terminate the program.



Refer to the sample output for formatting specifications.

Code constraints :
The given test cases fall under the following specifications:

1 ≤ ports[i] ≤ 65535

Sample test cases :
Input 1 :
127.0.0.1
80 443 8080

Port 80: CLOSED

Port 8080: OPEN
Input 2 :

Input 3 :

Port 25: TIMEOUT (Possibly closed)
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
# You are using Python
import socket
import sys
def is_val(ip):
    try:
        socket.inet_aton(ip)
        return True
    except socket.error:
        return False
def scan(ip,ports):
    if not is_val(ip):
        print("Invalid IP address.")
        return
    for port in ports:
        try:
            with socket.socket(socket.AF_INET,socket.SOCK_STREAM)as s:
                s.settimeout(2)
                result=s.connect_ex((ip,port))
                if result==0:
                    print(f"Port {port}: OPEN")
                elif result==10060:
                    print(f"Port {port}: TIMEOUT (Possibly closed)")
                else:
                    print(f"Port {port}: CLOSED")
        except socket.error:
            print(f"Port {port}:CLOSED")
if __name__=="__main__":
    ip=input().strip()
    port=list(map(int,input().split()))
    scan(ip,port)
                
                
                
                
                
                
```

# Single File Programming Question
**Problem Statement**



Alice, a network administrator, needs to deploy a local server that listens on multiple ports and verifies their accessibility. However, some ports might already be in use, preventing the server from binding to them. 



Your task is to implement a program that starts a server on given ports and checks whether a client can successfully connect to each of them. 



Help Alice ensure her network setup works correctly.



Note: In this compiler, port 8080 is already in use by default, which is why the server fails to start on port 8080, but the client can still connect to it. 

Input format :
The input consists of space-separated integers (ports[]), representing the port numbers to be used by the server.

Output format :
For each port, print "Server is listening on port X" if the server successfully binds to port X, otherwise "Failed to start the server on port X:[Errno 98] Address in use".

After attempting to start the servers, print "Connected to port X" if the client successfully connects to X.



Refer to the sample output for formatting specifications.

Code constraints :
The given test cases fall under the following specifications:

1024 ≤ ports[i] ≤ 65535

Sample test cases :
Input 1 :
8081 9090 10000
Output 1 :
Server is listening on port 8081
Server is listening on port 9090
Server is listening on port 10000
Connected to port 8081
Connected to port 9090
Connected to port 10000
Input 2 :
8080 9090
Output 2 :
Failed to start server on port 8080: [Errno 98] Address in use
Server is listening on port 9090
Connected to port 8080
Connected to port 9090
Note :
The program will be evaluated only after the “Submit Code” is clicked.
Extra spaces and new line characters in the program output will result in the failure of the test case.

```python
# You are using Python
import socket
import threading
def handle_client(server_socket):
    try:
        conn,addr=server_socket.accept()
        conn.close()
    except:
        pass
def start_servers(ports):
    servers=[]
    for port in ports:
        try:
            server_socket=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
            server_socket.bind(('localhost',port))
            server_socket.listen(1)
            threading.Thread(target=handle_client,args=(server_socket,),daemon=True).start()
            servers.append(server_socket)
            print(f"Server is listening on port {port}")
        except socket.error as e:
            print(f"Failed to start server on port {port}:{e}")
    return servers
def check_client_connections(ports):
    for port in ports:
        try:
            client_socket=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
            client_socket.connect(('localhost',port))
            print(f"Connected to port {port}")
            client_socket.close()
        except socket.error:
            pass
def main():
    ports=list(map(int,input().split()))
    start_servers(ports)
    check_client_connections(ports)
if __name__=="__main__":
    main()

                
            
```
